DATE FRUIT CLASSIFICATION USING ANN

```python
# Importing Required Libraries / Packages
import pandas as pd
import numpy as np
```

```python
# Dropping Class and making it as output
x = df.drop("Class", axis = 1)
y = df["Class"]
```

As it can be seen that there are 7 categories in the o/p, that means there will be 7 neurons in output layer.

```python
# Encoding the Output from Strings to numeric
from sklearn.preprocessing import StandardScaler, LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
# Splitting data into train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state=42)
```

```python
y
```

```python
# Scaling Features Because there is a big differernce in the values of features.

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

ANN

```python
import torch 
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
```

```python
# Coverting Data Points to Tensors
x_train_tensor = torch.tensor(x_train_scaled, dtype = torch.float32) 
y_train_tensor = torch.tensor(y_train, dtype = torch.long) 


x_test_tensor = torch.tensor(x_test_scaled, dtype = torch.float32)
y_test_tensor = torch.tensor(y_test, dtype = torch.long)
```

```python
# Creating Dataset with tensors

train_df = TensorDataset(x_train_tensor,y_train_tensor)
test_df = TensorDataset(x_test_tensor,y_test_tensor)
```

```python
# Creating Data Loader to fetch data from tensors

train_loader = DataLoader(train_df, batch_size = 32, shuffle = True)
test_loader = DataLoader(test_df, batch_size = 32)
```

BUILDING MODEL

```python
class ANN(nn.Module):
    def __init__ (self):
        super(ANN, self).__init__()
        self.model = nn.Sequential(
            # 1st Hidden layer
            nn.Linear(x.shape[1],64),
            nn.ReLU(),
            # 2nd Hidden Layer
            nn.Linear(64,64),
            nn.ReLU(),
            # Output Layer
            nn.Linear(64,7)
        )
    def forward(self, x):
        return self.model(x)
    
```

```python
model = ANN()

# Loss and Optimization
criteria = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters())
```

```python
# Training the Model/Neural Network

epochs = 100
for epoch in range (epochs):
    model.train() # Model in training mode

    running_loss = 0.0
    for xb, yb in train_loader:
        optimizer.zero_grad()
        outputs = model(xb)
        loss = criteria(outputs, yb)
        loss.backward()
        optimizer.step()

        running_loss += loss
    train_loss = running_loss/len(train_loader)
    print(f"epoch = {epoch+1}/{epochs}, loss = {train_loss}")
```

```python
# Model Evaluation 
model.eval()
total = 0
correct = 0

with torch.no_grad():# No gradients to be calculated
    for xb,yb in test_loader:
        outputs = model(xb)
        actul_val, predicted = torch.max(outputs, 1)# Returns Max value in tensors and the index of max value
        correct +=(predicted == yb).sum().item()
        total += yb.size(0) # Actual Samples in each batch
print("total values:", total)
print("Correct:", correct)
```

```python
Accuracy = correct/total
print("Accuracy:", Accuracy*100)
```
