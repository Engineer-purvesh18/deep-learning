## TEXT PRE PROCESSING

### 1. CONVERTING EVERYTHING TO LOWERCASE

```python
df_1['review'] = df_1['review'].str.lower()
```

### 2. Removing URLS

```python
# Removing URLS using Regex
import re

def remove_url(text):
    text = re.sub(r"http\S+" ,"", text) # (Pattern, replacment, string)
    return text
df_1['review'] = df_1['review'].apply(remove_url)
```

### 3. Removing Punctions

```python
def remove_punct(text):
    text = re.sub(r"[A-Za_z0-9\s]" ,"", text) # (Pattern, replacment, string) A-Z a_z 0-9 \s
    return text
    

df_1['review'] = df_1['review'].apply(remove_punct)
```

### 4. Removing HTML Tags

```python
def remove_html(text):
    text = re.sub(r"<.*?>" ,"", text) # (Pattern, replacment, string) 
    return text
    

df_1['review'] = df_1['review'].apply(remove_html)
```
